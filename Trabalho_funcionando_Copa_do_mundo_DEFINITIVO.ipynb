<br><span style="font-family:Verdana, Geneva, Tahoma, sans-serif;"><span style="font-weight: bold;">1</span> - Importando a biblioteca <span style="font-weight: bold;">basedosdados</span>, na qual possui a base de dados escolhida, ela foi previamente instalada na máquina. </span>

```python
import basedosdados as bd
```

<span style="font-family:Verdana, Geneva, Tahoma, sans-serif;"><span style="font-weight: bold;">2</span> – Atribuindo o conteúdo da base de dados a variável <span style="font-weight: bold;">df</span> . </span>

```python
df = bd.read_table(dataset_id='world_fifa_worldcup',
table_id='matches',
billing_project_id="projetoaulasegunda-383313")        
```

<span style="font-family:Verdana, Geneva, Tahoma, sans-serif;"> <span style="font-weight: bold;">3</span> – Processamento da aplicação, na qual o usuário realiza a busca de informações dos jogos da copa do mundo. Para que o programa retorne um resultado, é necessário digitar o nome do país no idioma inglês, visto que a base original foi feita neste idioma. </span>

```python
print("SEJA BEM VINDO AO SISTEMA DE PESQUISA DE JOGOS DA COPA DO MUNDO | ENTRE 1930 À 2018 | ")
pais=input("\nDigite o nome do país (em inglês) para obter informações sobre jogos em copas do mundo : ")
tipoTabela = input("Selecione uma das opções abaixo : \n1 - Para tabela completa \n2 - Para tabela filtrada(recomendado)\n\n")
if(tipoTabela=="1"):
    query="select * from `basedosdados.world_fifa_worldcup.matches` where upper(team_a_name) like upper ('%"+pais+"%') or upper(team_b_name) like upper('%"+pais+"%')"
    df=bd.read_sql(query,billing_project_id='projetoaulasegunda-383313')
elif(tipoTabela=="2"):
    query="select year,match_date,stadium_name,country_name,team_a_name,team_b_name,score,stage_name from `basedosdados.world_fifa_worldcup.matches` where upper(team_a_name) like upper ('%"+pais+"%') or upper(team_b_name) like upper('%"+pais+"%')"
    df=bd.read_sql(query,billing_project_id='projetoaulasegunda-383313')
else:
    print("Digite uma opção válida !!")
    query="select * from `basedosdados.world_fifa_worldcup.matches` where upper(team_a_name) like upper ('%tavazio%') or upper(team_b_name) like upper('%tavazio%')"
    df=bd.read_sql(query,billing_project_id='projetoaulasegunda-383313')
    
if((df.shape[0]==0) and (tipoTabela=="1" or tipoTabela=="2")):
    print("A consulta não obteve resultados!")
```

<span style="font-family:Verdana, Geneva, Tahoma, sans-serif;"><span style="font-weight: bold;">4</span> – Caso haja retorno do valor pesquisado, o programa mostra na tela o resultado em tabela, caso contrário, ocorre uma mensagem informando que não houve retorno da pesquisa. </span>

```python
if(df.shape[0]!=0):
    display(df)
```

<span style="font-family:Verdana, Geneva, Tahoma, sans-serif;"><span style="font-weight: bold;">5</span> – Caso haja retorno na pesquisa do usuário, a aplicação habilita a opção de fazer o download da tabela em excel, se assim desejado. </span>

```python
if(df.shape[0]!=0):
    FazerDownload = input("Deseja fazer o download do arquivo?\n 1 - Para sim\n 2 - Para não\n")
    if(FazerDownload=="1"):
        nomeArquivo = input("Digite o nome do arquivo a ser salvo\n")
        if(nomeArquivo!=""):  
            df.to_excel('C:/Users/usuario/Downloads/'+nomeArquivo+".xlsx",nomeArquivo+'.xlsx')
            print("Download feito com sucesso !!! \n Salvo em : C:/Users/usuario/Downloads/")
        else:
            print("O nome do arquivo não pode ser vazio !")   
    elif(FazerDownload=="2"):
        print("Download não solicitado pelo usuário!!!")
    else:
        print("Selecione uma opção válida !!")
else:
    print("Não houve retorno na consulta para gerar arquivo de download !")
```
